```python
import pandas as pd
```

```python
weatherdf = pd.read_csv('weatherdata.csv',  keep_default_na=True, sep=',\s+', delimiter=',')
weatherdf.shape
```

```python
weatherdf.head()
```

```python
weatherdf["new_dateframe"] = pd.to_datetime(weatherdf["datetime"], unit="s")
```

```python
weatherdf.head()
```

```python
weatherdf["hour"]= weatherdf["new_dateframe"].dt.hour
weatherdf["date"]= weatherdf["new_dateframe"].dt.date
weatherdf.head()
```

```python
weatherdf = weatherdf.drop(['id','w_descripti','datetime','day','new_dateframe'], axis = 1)
weatherdf.head()
```

```python
bikedf = pd.read_csv('bikesdata.csv',  keep_default_na=True, sep=',\s+', delimiter=',')
```

```python
bikedf.shape
```

```python
bikedf.head()
```

```python
bikedf["new_dateframe"] = pd.to_datetime(bikedf["day"], unit="ns")
bikedf["hour"]= bikedf["new_dateframe"].dt.hour
bikedf["date"]= bikedf["new_dateframe"].dt.date
bikedf.head()
```

```python
bikedf = bikedf.drop(['last_update','day','week','status', 'new_dateframe','HOUR(`day`)'], axis = 1)
bikedf.head()
```

```python
mergedf=pd.merge(bikedf,weatherdf,on=['hour','date'], how = "left")
mergedf.head()
```

```python
mergedf.shape
```

```python
mergedf = mergedf.dropna()
```

```python
mergedf.shape
```

```python
mergedf=mergedf.drop(['date'], axis = 1)
```

```python
mergedf.shape
```

```python
mergedf.dtypes
```

```python
categorical_columns = mergedf[['weather','week']].columns
for column in categorical_columns:
    mergedf[column] = mergedf[column].astype('category')

mergedf.dtypes
```

```python
mergedf.select_dtypes(['category']).describe().T
```

```python
mergedf= pd.get_dummies(mergedf, columns=['weather'], prefix = ['weather'])
```

```python
import pickle
for i in mergedf.id.unique():
    mergedf_id = mergedf.loc[mergedf["id"] == i]
    numeric_columns = mergedf_id.select_dtypes(['int64','float64','uint8']).columns
    numeric_columns = numeric_columns.drop(['available_bikes'])
    X = mergedf_id[numeric_columns]
    y = mergedf_id.available_bikes
    random_forest = RandomForestRegressor(max_depth=10, random_state=0, n_estimators=500)
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
    random_forest.fit(X_train,y_train)
    y_predicted = random_forest.predict(X_test)
    # printMetrics(y_test,y_predicted)
    # print(y_predicted)
    # print(y_predicted)
    pkl_filename = str(i) + ".pkl"  
    with open(pkl_filename, 'wb') as file:
        pickle.dump(random_forest, file)
    actual_vs_predicted_multiplelinreg1 = pd.concat([y_test,pd.DataFrame(y_predicted, columns=['Predicted'],index=y_test.index)],axis=1)
#     print(actual_vs_predicted_multiplelinreg1)
    print(i)
    print(printMetrics(y_test,y_predicted))
```
